```python
#@title Imports
!pip install -q symbulate
from symbulate import *

# !pip install tabulate
from tabulate import tabulate

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 5

mu = 0

alpha0 = 0.05

sigma0 = 1

sigma1 = 2

def var_known_mean(xs):
  return sum([(x - mu) ** 2 for x in xs]) / n

P0 = Normal(mu, sd = sigma0) ** n

Sigmahat2_H0 = RV(P0, var_known_mean)

sigmahat2_H0 = Sigmahat2_H0.sim(10000)

c = sigmahat2_H0.quantile(1 - alpha0)

sigmahat2_H0.plot()

Gamma(shape = n / 2, rate = 0.5 * n / sigma0 ** 2).plot()
plt.axvline(x = c, color = 'orange');
plt.axvspan(c, 100, alpha = 0.5, color = 'gray');

c, sigma0 ** 2 * ChiSquare(n).quantile(1 - alpha0) / n
```

```python
(n * sigmahat2_H0 / sigma0 ** 2).plot()
ChiSquare(n).plot()

ChiSquare(n).quantile(1 - alpha0)
```

```python
P1 = Normal(mu, sd = sigma1) ** n

Sigmahat2_H1 = RV(P1, var_known_mean)

sigmahat2_H1 = Sigmahat2_H1.sim(10000)

sigmahat2_H1.plot()
Gamma(shape = n / 2, rate = 0.5 * n / sigma1 ** 2).plot()
plt.axvline(x = c, color = 'orange');
plt.axvspan(c, 100, alpha = 0.5, color = 'gray');


sigmahat2_H1.count_gt(c) / 10000
```
