```
from symbulate import *
%matplotlib inline

from matplotlib import pyplot as plt
```

```
P = BoxModel([1, 2, 3, 4], size = 2, replace = True)
```

```
P.sim(10)
```

```
X = RV(P, sum)

Y = RV(P, max)
```

```
omega = (3, 2)

X(omega), Y(omega)
```

```
y = Y.sim(100)
y
```

```
y.tabulate()
```

```
y.tabulate(normalize=True)
```

```
y.plot('rug')
```

```
y.plot()
```

```
y = Y.sim(16000)
y.plot()
y.tabulate()
```

```
x = X.sim(16000)
x.plot()
x.tabulate()
```

```
y.sum() / 16000
```

```
y.sum() / y.count()
```

```
y.mean()
```

```
x.mean()
```

```
(Y == 3).sim(16000).tabulate()
```

```
xy = (X & Y).sim(100)
xy
```

```
xy.plot()
```

```
xy.plot(jitter = True)
```

```
xy.plot('tile')
```

```
xy = (X & Y).sim(16000)
xy.plot('tile')
xy.tabulate()
```

```
U = RV(P)
U.sim(10)
```

```
(U & X & Y).sim(10)
```

```
U1, U2 = RV(P)
(U1 & U2 & X & Y).sim(10)
```

```
P = DiscreteUniform(a = 1, b = 4)
U = RV(P)

U.sim(10000).plot()
```

```
P = DiscreteUniform(a = 1, b = 4) ** 2
X = RV(P, sum)
Y = RV(P, max)

(X & Y).sim(10000).plot(['tile', 'marginal'])
```
